```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

# FEATURES used for training — adjust if needed
features = [
    'grid','points','laps','milliseconds','fastestLap','fastestLapTime','fastestLapSpeed',
    'year','round','circuitId','position_qualifying','qual_vs_grid','driver-experiance',
    'constructor_avg_points','driver_avg_points','circuit_avg_time'
]
```

```python
def prepare_race_data(df, race_id=None, year=None, round_num=None, features=features):
  
    if race_id is not None:
        race_df = df[df['raceId'] == race_id].copy()
    else:
        race_df = df[(df['year'] == year) & (df['round'] == round_num)].copy()
    
    if race_df.empty:
        raise ValueError("No rows found for that race. Check raceId/year+round.")
    
    
    for col in features:
        if col not in race_df.columns:
            raise ValueError(f"Feature column missing from df: {col}")
    
  
    X_race = race_df[features].copy()
    X_race = X_race.fillna(0)                 
    X_race = X_race.astype(float, errors='ignore')
    

    context = race_df[['driverRef','constructorRef','grid','positionOrder']].reset_index(drop=True)
    return race_df.reset_index(drop=True), context, X_race
```

```python
def predict_race(race_id=None, year=None, round_num=None):
    race_df, context, X_race = prepare_race_data(df, race_id=race_id, year=year, round_num=round_num)
    preds = rf.predict(X_race)
    
    # Construct DataFrame of results
    out = context.copy()
    out['predicted_position'] = preds.round(0).astype(int)  
    out['predicted_position_float'] = preds                 
    out['actual_position'] = out['positionOrder']
    out['abs_error'] = (out['actual_position'] - out['predicted_position_float']).abs()
    
    
    out_sorted_by_pred = out.sort_values('predicted_position_float').reset_index(drop=True)
    out_sorted_by_actual = out.sort_values('actual_position').reset_index(drop=True)
    
    # per-race metrics
    mae_race = mean_absolute_error(out['actual_position'], out['predicted_position_float'])
    mse_race = mean_squared_error(out['actual_position'], out['predicted_position_float'])
    
    return {
        'race_df': race_df,
        'context': out,
        'pred_sorted': out_sorted_by_pred,
        'actual_sorted': out_sorted_by_actual,
        'mae': mae_race,
        'mse': mse_race
    }
```

```python
res = predict_race(year=2023, round_num=1)   # example: 2023 round 1
print("Race MAE:", res['mae'])
print("Race MSE:", res['mse'])
```

```python
def show_podium(res, top_n=3):
    pred_podium = res['pred_sorted'].head(top_n)[['driverRef','constructorRef','predicted_position_float','predicted_position']]
    actual_podium = res['actual_sorted'].head(top_n)[['driverRef','constructorRef','actual_position']]
    print("Predicted Podium:")
    display(pred_podium)
    print("Actual Podium:")
    display(actual_podium)

show_podium(res, top_n=3)
```

```python
def plot_pred_vs_actual(res):
    dfc = res['context'].copy().sort_values('actual_position')
    idx = np.arange(len(dfc))
    width = 0.35

    plt.figure(figsize=(10,6))
    plt.bar(idx - width/2, dfc['actual_position'], width=width, label='Actual')
    plt.bar(idx + width/2, dfc['predicted_position_float'], width=width, label='Predicted')
    plt.xticks(idx, dfc['driverRef'], rotation=90)
    plt.xlabel("Driver")
    plt.ylabel("Position (lower is better)")
    plt.title("Actual vs Predicted Positions for Race")
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_pred_vs_actual(res)
```

```python
def show_big_mistakes(res, top_n=10):
    dfc = res['context'].copy().sort_values('abs_error', ascending=False).head(top_n)
    display(dfc[['driverRef','constructorRef','grid','actual_position','predicted_position_float','abs_error']])

show_big_mistakes(res, top_n=10)
```
